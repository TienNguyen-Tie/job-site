```python
import requests
import pandas as pd
from datetime import date
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```python
#Time by Days
today = date.today()
```

```python
#Total Jobs of CareerBuilder
url = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-vi.html"
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs = soup.find("div", class_="ais-stats").find("span").get_text()
print(total_jobs)
```

```python
#Jobs by Industry
full_ind = []
full_no_of_jobs = []

industries = soup.find_all('div', class_="ais-refinement-list")[1].find_all("li")

for industry in industries:
    ind = industry.find_all("span")[0].get_text()
    no_of_jobs = industry.find_all("span")[1].get_text()
    full_ind.append(ind)
    full_no_of_jobs.append(no_of_jobs)
```

```python
#Jobs by Senority
#Jobs Sinh Viên
url_sv = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-fe1-vi.html"

page = requests.get(url_sv)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs_sv = soup.find("div", class_="ais-stats").find("span").get_text()


#Jobs Tốt nghiệp
url_tn = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-fe2-vi.html"

page = requests.get(url_tn)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs_tn = soup.find("div", class_="ais-stats").find("span").get_text()


#Jobs executive
url_nv = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-fe3-vi.html"

page = requests.get(url_nv)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs_nv = soup.find("div", class_="ais-stats").find("span").get_text()


#Jobs Senior
url_cv = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-fe4-vi.html"

page = requests.get(url_cv)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs_cv = soup.find("div", class_="ais-stats").find("span").get_text()


#Jobs Manager
url_tp = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-fe5-vi.html"

page = requests.get(url_tp)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs_tp = soup.find("div", class_="ais-stats").find("span").get_text()


#Jobs Director
url_gd = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-fe11-vi.html"

page = requests.get(url_gd)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs_gd = soup.find("div", class_="ais-stats").find("span").get_text()


#Jobs BoD
url_bdh = "https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-fe12-vi.html"

page = requests.get(url_bdh)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

total_jobs_bdh = soup.find("div", class_="ais-stats").find("span").get_text()
```

```python
#Candidate Number
url_candidate = "https://careerbuilder.vn/vi/resume-search.html"

page = requests.get(url_candidate)
soup = BeautifulSoup(page.text, 'html.parser')
soup.prettify()

full_can_ind_dict = {}
result_dict = {}

can_inds = soup.find('div', class_='HeaderLine CateEmp').find_all('li')

for inds in can_inds:
    can_ind = inds.find('a').get_text()
    no_of_can_ind = inds.find('span').get_text()
    no_of_can_ind = no_of_can_ind[1:-1]
    full_can_ind_dict[f'{can_ind}']=no_of_can_ind

df1 = pd.DataFrame(full_can_ind_dict.items(), columns = ['Industry','Number of Candidates'])
print(df1)


# for x_can_ind in full_can_ind_dict.keys():
#     if x_can_ind in full_ind:
#       result_dict = full_can_ind_dict
        
#         print(x_can_ind)

```

```python
#Create DataFrame
CareerBuilder = {'Time': today,
        'Industry': full_ind,
        'No. of Jobs': full_no_of_jobs,
        'Total Jobs': total_jobs,
        'Intern': total_jobs_sv,
        'Graduate': total_jobs_tn,
        'Executive': total_jobs_nv,
        'Senior': total_jobs_cv,
        'Manager': total_jobs_tp,
        'Director': total_jobs_gd,
        'Board of Director': total_jobs_bdh}

df = DataFrame(CareerBuilder, columns= ['Time', 'Total Jobs', 'Intern', 'Graduate', 'Executive', 'Senior', 'Manager', 'Director', 'Board of Director', 'Industry', 'No. of Jobs'])

print (df)
```

```python
#Merge DataFrame
result = pd.merge(df,
                 df1[['Industry', 'Number of Candidates']],
                 on='Industry')
print(result)
```

```python
result.to_csv("careerbuilder.csv")
```
